### Identifying Outliers -

#### Using Inter Quartile Range -

```python
Data["Average_income"]
```

```python
Q1 = Data.Average_income.quantile(0.25)

Q2 = Data.Average_income.quantile(0.5)

Q3 = Data.Average_income.quantile(0.75)

print(f"The first quantile of Average_income is : {Q1} ")
print(f"The second quantile of Average_income is : {Q2} ")
print(f"The third quantile of Average_income is : {Q3} ")

```

##### Outliers for Average_income -

```python
Q1 = Data.Average_income.quantile(0.25)

Q3 = Data.Average_income.quantile(0.75)

IQR = Q3 - Q1
print(f"IQR = {IQR}")

#Define upper and lower bound for outliers

Lower_bound = Q1 - 1.5*IQR
Upper_bound = Q3 + 1.5*IQR

Outliers = [x for x in Data["Average_income"] if x < Lower_bound or x > Upper_bound]

print(f" Lower_bound : {Lower_bound} ")
print(f" Upper_bound : {Upper_bound} ")
print(f" Identified Outliers : {Outliers} ")
```

##### Outliers for Petrol_consumption -

```python
Q1 = Data.Petrol_Consumption.quantile(0.25)

Q3 = Data.Petrol_Consumption.quantile(0.75)

IQR = Q3 - Q1

print(f" IQR = {IQR} ")

# Define Lower and Upper bound -

Lower_bound = Q1 - 1.5*IQR
Upper_bound = Q3 - 1.5*IQR

#Outliers 

Outlier = [x for x in Data["Petrol_Consumption"] if x < Lower_bound or x > Upper_bound]

print(f" Lower_bound : {Lower_bound} ")
print(f" Upper_bound : {Upper_bound} ")
print(f"Ouliers : {Outliers} ")

```

```python
Q1 = Data.Paved_Highways.quantile(0.25)

Q3 = Data.Paved_Highways.quantile(0.75)

IQR = Q3 - Q1

print(f" IQR = {IQR} ")

# Define Lower and Upper bound -

Lower_bound = Q1 - 1.5*IQR
Upper_bound = Q3 - 1.5*IQR

#Outliers 

Outlier = [x for x in Data["Paved_Highways"] if x < Lower_bound or x > Upper_bound]

print(f" Lower_bound : {Lower_bound} ")
print(f" Upper_bound : {Upper_bound} ")
print(f"Ouliers : {Outliers} ")
```

```python
Q1 = Data.Petrol_tax.quantile(0.25)

Q3 = Data.Petrol_tax.quantile(0.75)

IQR = Q3 - Q1

print(f" IQR = {IQR} ")

# Define Lower and Upper bound -

Lower_bound = Q1 - 1.2 * IQR
Upper_bound = Q3 - 1.2 * IQR

#Outliers 

Outlier = [x for x in Data["Petrol_tax"] if x < Lower_bound or x > Upper_bound]

print(f" Lower_bound : {Lower_bound} ")
print(f" Upper_bound : {Upper_bound} ")
print(f"Ouliers : {Outliers} ")
```
